### Filtro
La selección de filas basada en una determinada condición se puede realizar con la indexación booleana. Esto utiliza los valores reales de los datos en el DataFrame en lugar de las etiquetas de fila/columna o las posiciones de índice.

```python
boroughs['Average Age'] > 39
```

Cuando quieras seleccionar las filas y ver todos los datos añade `boroughs[]` alrededor de tu función:

```python
boroughs[boroughs['Average Age'] > 39]
```

Puede combinar diferentes columnas utilizando los operadores `&`,`|` y `==`

```python
# AND (&)
boroughs[(boroughs['Average Age'] > 39) & (boroughs['Political control'] == 'Cons')]
```

```python
# OR (|)
boroughs[(boroughs['Political control'] == 'Lab') | (boroughs['Political control'] == 'Lib Dem')]
```

## 5. Fusión de Datos

Pandas tiene varias opciones diferentes para combinar o fusionar datos. La documentación https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html tiene muchos ejemplos.

Vamos a crear dos nuevos Dataframes para explorar esto: cities y cities2

```python
import pandas as pd

data = {'city':      ['London', 'Manchester', 'Birmingham', 'Leeds', 'Glasgow'], 
       'population': [999343, 2343895, 4323456, 1243565, 4959093],
       'area':       [1734.4, 235.6, 3621.3, 5454.0, 7643.4]}
ciudad = pd.DataFrame(data)


data2 ={'city':       ['Liverpool', 'Southampton'],
        'population': [44556, 77534],
        'area':       [144.4, 234.6]}
ciudad2 = pd.DataFrame(data2)
```

```python
ciudad
```

```python
ciudad2
```

Usa `append()` para combinar estos Dataframes:

```python
ciudad3 = ciudad.append(ciudad2, ignore_index=True)
```

```python
ciudad3
```

```python
data = {'city':      ['London', 'Manchester', 'Birmingham', 'Leeds', 'Glasgow'], 
       'density':    [3452, 4521, 9732, 3452, 5633]}
ciudad4 = pd.DataFrame(data)
```

```python
ciudad4
```

Se puede agregar una columna adicional con `.merge()` con una unión externa usando los nombres de las ciudades:

`how= 'outer'` Devuelve todas las filas de ambas tablas, une los registros de la izquierda que tienen claves coincidentes en la tabla de la derecha. Cuando no hay coincidencias de ninguna tabla, se devolverá NaN

```python
ciudad3=pd.merge(ciudad3, ciudad4, how='outer', sort=True, on='city')
ciudad3
```

Pandas usa `Matplotlib` como predeterminado para las visualizaciones.

El gráfico predeterminado es un gráfico de líneas que utiliza el índice para el eje x:

```python
boroughs['Employment rate (%)'].plot();
```

Para crear un gráfico que tenga más sentido para estos datos, consulte la documentación de todas las opciones: https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html

Para el ejemplo anterior, un histograma podría funcionar mejor. Puede cambiar el número de columnas: `bins` para obtener el resultado deseado:

```python
boroughs['Employment rate (%)'].plot.hist(bins=10);
```

cambia el tamaño del histograma con la opcion `figsize`

```python
#figsize(ancho,altura)
boroughs['Employment rate (%)'].plot.hist(bins=15,figsize=(10,5));
```

Dentro del comando de trazado puede seleccionar los datos directamente. El siguiente histograma muestra la tasa de empleo solo para las afueras de Londres ('Outer London'):

```python
boroughs['Employment rate (%)'][boroughs['Inner/Outer']=='Outer London'].plot.hist(bins=15,figsize=(10,5));
```

Para agregar la tasa de empleo para el centro de Londres ('Inner London'), repita el comando de trazado con una selección diferente de datos:

```python
boroughs['Employment rate (%)'][boroughs['Inner/Outer']=='Outer London'].plot.hist(bins=15,figsize=(10,5));
boroughs['Employment rate (%)'][boroughs['Inner/Outer']=='Inner London'].plot.hist(bins=15,figsize=(10,5));
```

El gráfico anterior es difícil de leer ya que los histogramas se han superpuesto. Puede solucionar esto cambiando los colores y haciéndolos transparentes.

Para agregar una leyenda, cada histograma debe asignarse a un eje de `ax`. Con `legend()` puede agregar una leyenda. Con `plt.xlabel()` también puede agregar una etiqueta para el eje x (esto funciona de manera similar para el eje y):

```python
ax = boroughs['Employment rate (%)'][boroughs['Inner/Outer']=='Outer London'].plot.hist(
    bins=15,figsize=(10,5),alpha=0.5,color='#8E44AD');
ax = boroughs['Employment rate (%)'][boroughs['Inner/Outer']=='Inner London'].plot.hist(
    bins=15,figsize=(10,5),alpha=0.5,color='#3498DB');
ax.legend(['Outer London','Inner London'])
plt.xlabel('Employment rate (%)');
```

### Seaborn

Seaborn es una biblioteca de visualización de datos de Python basada en matplotlib. Es un paquete de visualización fácil de usar que funciona bien con Pandas DataFrames.

A continuación se muestran algunos ejemplos que utilizan Seaborn.

Consulte esta documentación para obtener información sobre muchos gráficos que puede crear.

```python
import seaborn as sns
```

Veamos un gráfico de distribución usando `distplot`, que muestra una distribución de los datos.

Use la función `dropna()` para eliminar filas y columnas con valores Null/NaN:

```python
sns.distplot(boroughs['Population density (/ha)'].dropna());
```

Puede crear diagramas categóricos con `catplot`. Hay diagramas de dispersión categóricos, diagramas de distribución y diagramas de estimación. El parámetro `kind` selecciona la función a usar, por ejemplo box, violin, swarm, bar, stripplot y boxen.

La representación predeterminada en catplot() usa un diagrama de dispersión:

```python
sns.catplot(x='Turnout at local elections', y='Political control', data=boroughs);
```

Also try `kind="swarm"`, `kind="box"` or `kind="violin"`:

```python
sns.catplot(x='Median House Price', y='Name', kind='swarm', data=boroughs);
```

```python
sns.catplot(x='Employment rate (%)', y='Largest migrant population', kind="box", data=boroughs);
```

```python
sns.catplot(x='Turnout at local elections', y='Political control', kind="violin", data=boroughs);
```
