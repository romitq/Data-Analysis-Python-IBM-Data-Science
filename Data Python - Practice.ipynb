## 5. FusiÃ³n de Datos

Pandas tiene varias opciones diferentes para combinar o fusionar datos. La documentaciÃ³n https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html tiene muchos ejemplos.

Vamos a crear dos nuevos Dataframes para explorar esto: cities y cities2

```python
import pandas as pd

data = {'city':      ['London', 'Manchester', 'Birmingham', 'Leeds', 'Glasgow'], 
       'population': [999343, 2343895, 4323456, 1243565, 4959093],
       'area':       [1734.4, 235.6, 3621.3, 5454.0, 7643.4]}
ciudad = pd.DataFrame(data)


data2 ={'city':       ['Liverpool', 'Southampton'],
        'population': [44556, 77534],
        'area':       [144.4, 234.6]}
ciudad2 = pd.DataFrame(data2)
```

```python
ciudad
```

```python
ciudad2
```

Usa `append()` para combinar estos Dataframes:

```python
ciudad3 = ciudad.append(ciudad2, ignore_index=True)
```

```python
ciudad3
```

```python
data = {'city':      ['London', 'Manchester', 'Birmingham', 'Leeds', 'Glasgow'], 
       'density':    [3452, 4521, 9732, 3452, 5633]}
ciudad4 = pd.DataFrame(data)
```

```python
ciudad4
```

Se puede agregar una columna adicional con `.merge()` con una uniÃ³n externa usando los nombres de las ciudades:

`how= 'outer'` Devuelve todas las filas de ambas tablas, une los registros de la izquierda que tienen claves coincidentes en la tabla de la derecha. Cuando no hay coincidencias de ninguna tabla, se devolverÃ¡ NaN

```python
ciudad3=pd.merge(ciudad3, ciudad4, how='outer', sort=True, on='city')
ciudad3
```

El grÃ¡fico anterior es difÃ­cil de leer ya que los histogramas se han superpuesto. Puede solucionar esto cambiando los colores y haciÃ©ndolos transparentes.

Para agregar una leyenda, cada histograma debe asignarse a un eje de `ax`. Con `legend()` puede agregar una leyenda. Con `plt.xlabel()` tambiÃ©n puede agregar una etiqueta para el eje x (esto funciona de manera similar para el eje y):

```python
ax = boroughs['Employment rate (%)'][boroughs['Inner/Outer']=='Outer London'].plot.hist(
    bins=15,figsize=(10,5),alpha=0.5,color='#8E44AD');
ax = boroughs['Employment rate (%)'][boroughs['Inner/Outer']=='Inner London'].plot.hist(
    bins=15,figsize=(10,5),alpha=0.5,color='#3498DB');
ax.legend(['Outer London','Inner London'])
plt.xlabel('Employment rate (%)');
```
